```python
# 由于Unnamed缺失过多我们将它删除
data.drop(['Unnamed: 32'], axis=1, inplace=True)
data.info()  #查看数据
```

```python
# 数据集划分
# 数据集除了诊断和ID两个属性之外都为检测各项数据用于判断是否患病
# 其他30个变量都是由10个数字化细胞核的10个不同特征的均值、标准差和最大值构成
from sklearn.model_selection import train_test_split

X=data.drop(['id', 'diagnosis'], axis=1)
y=data['diagnosis']

train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 0.2, random_state = 10)
```

```python
# 各取值范围太大了所以要进行标准化将范围减小到[0,1]
```

```python
# min-max标准化
def min_max_normalize(x):
    return (x - x.min())/(x.max() - x.min())
```

```python
for col in data.columns[2:31]:
    data[col] = min_max_normalize(data[col])
```

```python
data.head()
```

```python
# 再进行一次数据划分
X=data.drop(['id', 'diagnosis'], axis=1)
y=data['diagnosis']

train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 0.2, random_state = 10)
```

```python
X.describe()
```

```python
# 各数值的范围都在[0,1]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
import seaborn as sns

neigh = KNeighborsClassifier()   #默认k值为5
neigh.fit(train_X, train_y)    #模型预测

y_pred = neigh.predict(test_X)    #性能评估

print("分类正确率：",round(neigh.score(test_X, test_y),4))   #返回浮点数x的四舍五入值

ax = sns.heatmap(confusion_matrix(test_y, y_pred), 
                 annot=True, fmt='d', 
                 xticklabels=["良性(B)","恶性(M)"], yticklabels=["良性(B)","恶性(M)"])
ax.set_ylabel('真实')
ax.set_xlabel('预测')
ax.set_title('混淆矩阵热力图')
```

```python
import numpy as np
cat=np.random.rand(30)
cat=np.array(cat).reshape(1, -1)
# 这是由于在新版的sklearn中，所有的数据都应该是二维矩阵，哪怕它只是单独一行或一列（比如前面做预测时，仅仅只用了一个样本数据）
cat
```

```python
neigh.predict(cat)
```

```python
# 查看近邻情况
plt.scatter(train_X.values[:,0],train_X.values[:,1],alpha=0.4)
plt.scatter(cat[:,0],cat[0,1],color='red',alpha=0.8)
```

```python
# 可以看出分类正确率还是很高的
```

```python
# 接下来我们来看一下不同的K值对准确率的影响
```

```python
K_scores = [KNeighborsClassifier(n_neighbors=k).fit(train_X, train_y).score(test_X, test_y) for k in range(1,10)]
K_scores
```

```python
# 绘图
plt.plot(range(1, 10), K_scores,linewidth=3)
plt.xlabel('K')
plt.ylabel('正确率')
plt.title('不同K值下模型预测正确率')
```

```python
K_scores_weights = [KNeighborsClassifier(n_neighbors=k,weights='distance').fit(train_X, train_y).score(test_X, test_y) for k in range(1,10)]
plt.plot(range(1, 10), K_scores_weights, 'g-.', linewidth=3)
plt.xlabel('K值')
plt.ylabel('正确率')
plt.title('不同K值下模型预测正确率')
```

```python
data_M=data[data['diagnosis']=='M']
data_B=data[data['diagnosis']=='B']
```

```python
plt.figure(figsize=(10,5))
plt.scatter(data_M['radius_mean'],data_M['texture_mean'],color='r',label='恶性')
plt.scatter(data_B['radius_mean'],data_B['texture_mean'],color='black',label='良性')
plt.legend(loc=1)
plt.xlabel("半径")
plt.ylabel("质地")
plt.show()
```

```python
# 半径和质地的二维图可以发现，良性主要集中在半径[0,0.4]和质地[0,0.8]，
# 而恶性虽然也有和良性重叠的地方，但是主要还是集中在右边
```

```python
# 我们在来观察一下其他参数的二维图
```

```python
plt.figure(figsize=(10,5))
plt.scatter(data_M['concavity_mean'],data_M['concave points_mean'],color='r',label='恶性')
plt.scatter(data_B['concavity_mean'],data_B['concave points_mean'],color='black',label='良性')
plt.legend(loc=1)
plt.xlabel("凹度")
plt.ylabel("凹点")
plt.show()
```

```python
# 这是凹度与凹点的二维图
# 可以明显的看出凹度凹点越大，患病几率就会越大
```

```python
# 再来看最后一组二维图
# 光滑度和致密性
```

```python
plt.figure(figsize=(10,5))
plt.scatter(data_M['smoothness_mean'],data_M['compactness_mean'],color='r',label='恶性')
plt.scatter(data_B['smoothness_mean'],data_B['compactness_mean'],color='black',label='良性')
plt.legend(loc=1)
plt.xlabel("光滑度")
plt.ylabel("致密性")
plt.show()
```

```python
# 两者的对比发现重合度还是比较高的
# 但是致命性越高患病几率就越大
```
